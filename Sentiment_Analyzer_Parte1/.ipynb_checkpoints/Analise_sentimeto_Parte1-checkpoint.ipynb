```python
# Importa√ß√µes necess√°rias
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
from pymongo import MongoClient
import pandas as pd
import json
```

<p>A seguir ser√° preciso inserir as credenciais de seguran√ßa que mencionei para poder coletar os tweets.<br>
Por quest√µes de seguran√ßa eu executei o c√≥digo e removi as minhas.</p>

```python
# Autenticando as chaves
auth = OAuthHandler(consumer_key, consumer_secret)
```

```python
auth.set_access_token(access_token, access_token_secret)
```

```python
# Lista vazia que receber√° os tweets, caso voc√™ queira armazen√° los em um banco de dados noSQL como o mongodb fique a vontade
# o processo √© o mesmo.
tweets_list = []
```

```python
# Definindo a quantidade de tweets que quero armazenar.
limit = 500
```

```python
# Criando uma classe para capturar o stream de dados do Twitter 
class MyListener(StreamListener):
    global count
    def on_data(self, dados):
        count = 0
        tweet = json.loads(dados)
        created_at = tweet["created_at"]
        text = tweet["text"]
        obj = {"created_at":created_at,"text":text}
        
        # Eu aconselho a fazer o c√≥digo igual ao meu, pelo menos a parte de cima no qual coleto os tweets pois as vezes ocorre alguns
        # bugs
        
        
        # Adicionando os tweets que est√£o em um dicion√°rio a lista vazia 
        while len(tweets_list) < limit:
            tweets_list.append(obj)
            print("Quantidade tweets coletados", len(tweets_list))
            print (obj)
            count +=1
            return True
        else:
            pass
```

```python
# Instanciando a classe MyListener
Mylistener = MyListener()
```

```python
# Chamando a fun√ß√£o Stream passando as chaves autenticadas e a classe com as instru√ß√µes de como coletar os tweets
myStream = Stream(auth, listener= Mylistener)
```

```python
# Passando as palavras chaves que irei usar para filtrar os tweets
keywords = ["data science", 'data analytics', 'power bi']
```

<p>Importante: A seguir irei iniciar a coleta dos tweets fazendo uso das palavras chaves, ele somente armazenar√° os 500 que determinei<br>
entretanto o programa continuar√° coletando, para interromper o processo clique no quadrado na barra de ferramentas. A mensagem em vermelho que aparecer√° n√£o indica um erro mas sim que o processo foi subitamente parado.</p>

```python
# Coletando os tweets
myStream.filter(track= keywords)
```

```python
# Ap√≥s realizar a coleta √© importante desconectar do Streaming
myStream.disconnect()
```
